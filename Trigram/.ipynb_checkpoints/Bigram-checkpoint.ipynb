```python
import torch
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/companies.csv")
```

```python
df.shape
```

```python
data = df.iloc[:,0]
```

```python
names = (set(data.tolist()))
```

```python
with open("sanitized_companies.txt", "w") as f:
    for name in names:
        if pd.notna(name) and name.isascii():
            sanName = re.sub(r'\(.*?\)|\[.*?\]|\{.*?\}|-.*|:.*|–.*|\|.*|[,，!$%&+\s\'’",./_@]', '', name)
            f.write(str(sanName) + "\n");
```

```python
#default loading of the sanitized companies list:
```

```python
data = [item.lower() for item in open("./dataset/sanitized_companies.txt", "r").read().split('\n')]
```

```python
engWords = open("./dataset/better_words.txt", "r").read().split('\n')
```

```python
chars = sorted((set(''.join(data))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
```

```python
len(engWords), len(data)
```

```python
#Actual Model making starts now
```

```python
rawCounts = torch.zeros((37,37), dtype=torch.int32)
```

```python
for name in data:
    aug = '.' + str(name).lower() + '.' #augmenting . to hallucinate the start and the end token
    for ch1, ch2 in zip(aug[0:], aug[1:]):
        rawCounts[stoi[ch1], stoi[ch2]] += 1
```

```python
for word in engWords:
    aug = '.' + word.lower() + '.'
    for ch1, ch2 in zip(aug[0:], aug[1:]):
        rawCounts[stoi[ch1], stoi[ch2]] += 3 #3x weightage for generating english names
```

```python
%matplotlib inline

plt.figure(figsize=(16,16))
plt.axis('off')
plt.imshow(rawCounts, cmap = 'Greens')
for i in range(36):
    for j in range(36):
        label = itos[i] + itos[j]
        plt.text(j, i, label, ha = "center", va = "bottom", color = 'grey')
        plt.text(j, i, rawCounts[i,j].item(), ha = "center", va = "top", color = 'grey', size = '6')
```

```python
rowSum = rawCounts.sum(dim=1, keepdim = True)
probTable = rawCounts/rowSum
```

```python
generator = torch.Generator().manual_seed(43)
```

```python
generations = []
for i in range(10):
    currTokId = stoi['.']
    currString = ''
    while True:
        nextTokId = torch.multinomial(probTable[currTokId], num_samples=1,replacement=True, generator = generator).item()
        nextTok = itos[nextTokId]
        if nextTok == '.': break;
        currString+= nextTok
        currTokId = nextTokId
    generations.append(currString)
generations
```

```python
#Since this Approach is not very accurate we need to take a neural network Approach. 
#But before doing that let me show you how to calculate the loss of this model.

#Loss can be calculated using Likelihood Estimations (MLE)
#We can negate the log likelihood estimation to get the actual loss
```

```python
totalSum = 0
n = 0
for name in data:
    aug = '.' + str(name).lower() + '.'
    for ch1, ch2 in zip(aug[0:], aug[1:]):
        totalSum += torch.log(probTable[stoi[ch1], stoi[ch2]])
        n+=1
print(f'Total log likelihood: {totalSum =}')
negLogLikelihood = -totalSum;
print(f'Total neg log likelihood: {negLogLikelihood =}')
print(f'Avg neg Log likelihood: {negLogLikelihood/n}')
```

```python
#Now let us shift to a better way of making model. Using One-hot encoding and making a neural network.
```

```python
#Now let's remake this model using neural networks and one-hot encoding
```

```python
#First let's make our training sets
xs, ys = [], []
for name in data:
    aug = '.' + str(name).lower() + '.'
    for ch1, ch2 in zip(aug[0:], aug[1:]):
        xs.append(stoi[ch1]);
        ys.append(stoi[ch2]);
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
xenc
```

```python
import torch.nn.functional as F
```

```python
# implementing 2 layer network from scratch to do this
# let's have two hidden layers
#  '- 16 neurons each
# input matrix will be 1 x 37
# W1 matrix will be 37 x 16 and the bias matrix b1 will be 1 x 16
# W2 matrix will be  1 x 16 and the bias matrix b2 will be 1 x 16
# W3 will just be normalization matrix, it will be 16 x 37 matrix that will bring back the dimensions. 
# We will also optimize this. We won't have any bias here.
```

```python
# initializing all the weights and biases to random values:
G2 = torch.Generator().manual_seed(2)
W1 = torch.randn((37,16),generator = G2, requires_grad=True)
B1 = torch.randn((1, 16), generator = G2, requires_grad = True)
W2 = torch.randn((16,16),generator = G2, requires_grad=True)
B2 = torch.randn((1, 16), generator = G2, requires_grad = True)
W3 = torch.randn((16, 37), generator = G2, requires_grad = True)
B3 = torch.zeros((1, 37), requires_grad=True)
W1.grad = None
W2.grad = None
B1.grad = None
B2.grad = None
W3.grad = None
B3.grad = None
```

```python
# Forward Pass
xenc = F.one_hot(xs, num_classes=37).float()
s1 = (xenc @ W1) + B1
s2 = (s1 @ W2) + B2
logits = s2 @ W3 + B3
# logits[0] - for preview
counts = logits.exp()
probs = counts/counts.sum(dim=1, keepdim=True)
loss = - probs[torch.arange(len(ys)),ys].log().mean()
loss
```

```python
# Backward Pass
loss.backward()
```

```python
# Update
W1.data += -0.1*W1.grad
W2.data += -0.1*W2.grad
W3.data += -0.1*W3.grad
B1.data += -0.1*B1.grad
B2.data += -0.1*B2.grad
B3.data -= 0.1*B3.grad
```

```python
#### Piecing it all together in gradient Decent:

for i in range(1000):
    xenc = F.one_hot(xs, num_classes=37).float()
    s1 = torch.tanh(xenc @ W1 + B1)
    s2 = torch.tanh(s1 @ W2 + B2)
    logits = s2 @ W3 + B3
    logits = logits - logits.max(dim = 1, keepdim = True).values
    counts = logits.exp()
    probs = counts/counts.sum(dim=1, keepdim=True)
    loss = - probs[torch.arange(len(ys)),ys].log().mean()
    # loss = F.cross_entropy(logits, ys)
    for p in [W1, W2, W3, B1, B2, B3]:
        p.grad = None
    # Backward Pass
    loss.backward()

    # Update
    lr = 0.5

    with torch.no_grad():
        W1 -= lr * W1.grad
        W2 -= lr * W2.grad
        W3 -= lr * W3.grad
        B1 -= lr * B1.grad
        B2 -= lr * B2.grad
        B3 -= lr * B3.grad

    print(loss.data)
```

```python
# Generation:

G3 = torch.Generator().manual_seed(43)
feed = '.'
currGen = ''
output = ''
while currGen != '.':
    xenc = F.one_hot(torch.tensor(stoi[feed]), num_classes=37).float()
    s1 = torch.tanh(xenc @ W1 + B1)
    s2 = torch.tanh(s1 @ W2 + B2)
    logits = s2 @ W3
    logits = logits - logits.max(dim = 1, keepdim = True).values
    counts = logits.exp()
    probs = counts/counts.sum(dim=1, keepdim=True)
    currGen = itos[torch.multinomial(probs, 1).item()]
    output+=currGen
    feed = currGen
print(output)
```
